```python
data.drop("Unnamed: 0",axis=1)
```

Gender Distribution

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data = data, x="Gender")
ax.bar_label(ax.containers[0])
plt.title('Gender Distribution')
plt.show()
```

```python
#from the above chart we have analysed that:
#the number of females in the data is more likely is more than the number of males
```

```python
gb = data.groupby('ParentEduc').agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb, annot=True)
plt.title("Relatinship between Parent's Education and Student's Score")
plt.show()
```

```python
#from the above chart we have concluded that the education of the parents have a good impact in student's score
```

```python
gb1 = data.groupby('ParentMaritalStatus').agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
sns.heatmap(gb1, annot=True)
plt.title("Relatinship between Parent's Education and Student's Score")
plt.show()
```

```python
#from the above chart we have concluded that their is no/negligible impact on the student's score
#due to their parent's marital status
```

Distribution of Ethnic Groups

```python
groupA=data.loc[(data['EthnicGroup']=='group A')].count()
groupB=data.loc[(data['EthnicGroup']=='group B')].count()
groupC=data.loc[(data['EthnicGroup']=='group C')].count()
groupD=data.loc[(data['EthnicGroup']=='group D')].count()
groupE=data.loc[(data['EthnicGroup']=='group E')].count()

l = ["Group A","Group B","Group C","Group D","Group E"]
mlist=[groupA['EthnicGroup'],groupB['EthnicGroup'],groupC['EthnicGroup'],groupD['EthnicGroup'],groupE['EthnicGroup']]
plt.pie(mlist, labels=l, autopct="%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```
